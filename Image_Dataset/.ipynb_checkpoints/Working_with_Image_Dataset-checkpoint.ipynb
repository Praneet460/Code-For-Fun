### Working With Image Dataset

#### 1. Import Required Libraries

```python
import numpy as np
import os
import matplotlib.pyplot as plt

from IPython.core.display import Image, display
%matplotlib inline
```

#### Data Preprocessing

```python
# Create a list of file names
# set current working directory
cwd = os.getcwd()
# path of images
image_path = os.path.join(cwd, 'images')
# create a list to store file names
images = os.listdir(image_path)
# print the list
print(images)
print("\n")
print("No. of images = ", len(images))
```

```python
# Display the images
for f in images[0:6]:
    display(Image(filename = os.path.join(image_path, f), width = 70, height = 50))
```

```python
# ratio of red portion and green portion in the images
red_portion = []
green_portion = []
for image in images:
    img = plt.imread(os.path.join(image_path, image))
    reds = img[:,:,0]
    red_portion.append(np.sum(reds))
    greens = img[:,:,1]
    green_portion.append(np.sum(greens))
    
ratio = np.array(red_portion, dtype = float) / np.array(green_portion, dtype = float)
print(ratio)
print("\n")
print("Size of ratio array is ", len(ratio))
```

```python
# Plot 
plt.figure(1, figsize = (10, 8))
plt.subplot(221)
plt.plot(range(len(red_portion)), red_portion, 'ro', label = "red_portion")
plt.plot(range(len(green_portion)), green_portion, 'go', label = "green_portion")
plt.title("Green and Red Portion")

plt.subplot(222)
plt.plot(range(len(red_portion)), red_portion, c = 'red', label = "red_portion")
plt.plot(range(len(green_portion)), green_portion, c = 'green', label = "green_portion")
plt.title("Red and Green Portion")

plt.subplot(223)
plt.plot(range(len(ratio)), ratio)
plt.title("Red and Green Ratio Line Chart")

plt.subplot(224)
plt.bar(range(len(ratio)), ratio, color = 'pink')
plt.title("Red and Green Ratio Bar Chart")

plt.show()
```
